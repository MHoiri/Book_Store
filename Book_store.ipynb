```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python

url='http://books.toscrape.com/catalogue/category/books/mystery_3/index.html'

response = requests.get(url)
```

```python
soup =BeautifulSoup(response.text)
```

```python
Book = soup.find_all('article', class_='product_pod')
Book
```

```python
for art in Book:
    h_3= art.find('h3')
    a=h_3.find('a')
    name=a.attrs['title']


    price= art.find('div', attrs={'class':'product_price'})
    pri=price.find('p',attrs={"class":'price_color'}).text
    pri=float(pri[2:])


    rate=art.find('p', class_='star-rating')
    rate=rate['class'][1]
    if rate=='Four':
        rate=4
    elif rate=='One':
        rate=1
    elif rate=='Two':
        rate=2
    elif rate=='Three':
        rate=3

        
    print(f'Name: {name} \n Price: {pri} \n Rate: {rate}') 
```

```python
for i in range(49):
    li=soup.find("li",class_= "next")
    next_page=li.find("a").attrs
    next_url= next_page['href']
    next_url="https://books.toscrape.com/catalogue/"+ next_url
    
    url=next_url
    response = requests.get(url)
    soup =BeautifulSoup(response.text)
    Book = soup.find_all('article', class_='product_pod')

    for art in Book:
        h_3= art.find('h3')
        a=h_3.find('a')
        name=a.attrs['title']


        price= art.find('div', attrs={'class':'product_price'})
        pri=price.find('p',attrs={"class":'price_color'}).text
        pri=float(pri[2:])


        rate=art.find('p', class_='star-rating')
        rate=rate['class'][1]
        if rate=='Four':
            rate=4
        elif rate=='One':
            rate=1
        elif rate=='Two':
            rate=2
        elif rate=='Three':
            rate=3

            
        print(f'Name: {name} \n Price: {pri} \n Rate: {rate}') 
        
        with open('Book Data','a',encoding="utf-8",newline='') as f:
            writer= csv.DictWriter(f,fieldnames=['Book Name','Price','Rate'])
            writer.writeheader()
            writer.writerow(
                {
                    'Book Name':name,
                    'Price':pri,
                    'Rate':rate

                 }
            )
    
        
```
